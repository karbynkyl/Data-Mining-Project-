# Optimization

Linear programming

$z = 40x + 30y$

Subject to:

$2x + y \leq 100$

$x + y \leq 80$

$x \leq 50$

```python
!pip install pulp
```

```python
from pulp import LpMaximize, LpProblem, LpVariable
```

```python
# create an instance of a model
# with its objective function set as maximization
model = LpProblem('Profi_maximizer', LpMaximize)

# we create variables that will represent the variables
# from the function that is to be optimized
# and add any constrains that are related to their nature
x = LpVariable('P1', lowBound = 0)
y = LpVariable('P2', lowBound = 0)

# we add the main function
model += 40 * x + 30 * y

# after the function that we are optimizing
# we add all the other functions that map
# its constraints
model += 2 * x + y <= 100
model += x + y <= 80
model += x <= 50

# we call the solver to provide us with the solution
model.solve()

print(f"x = {x.value()}")
print(f"y = {y.value()}")

print(f"z = {model.objective.value()}")
```

```python
from scipy.optimize import linprog

# Maximize:
# z = 3x1 + 2x2
# x1 + x2 <= 4
# x1 <= 2
# x2 <= 3
# -------------------------------------------------
# x1, x2 >= 0; this is a boundary, not a constraint

# linprog does minimization, we change the sign before our coeficients to perform maximization
C = [-3, -2]

# Left-hand side
A = [
    [1, 1],
    [1, 0],
    [0, 1]
]

# Right-hand side
B = [4, 2, 3]

# boundaries
bounds = [(0, None), (0, None)] # eliminates negative values

# model
model = linprog(C, A_ub = A, b_ub = B, bounds = bounds, method = 'simplex')

print(model)
```

```python
fig, ax  = plt.subplots(figsize = (6, 6))

x = np.linspace(0, 5, 400)
y1 = 4 - x
y2 = np.full_like(x, 3)
y3 = np.full_like(x, 2)

x_fill = [0, 0, 1, 2, 2]
y_fill = [0, 3, 3, 2, 0]

ax.fill(x_fill, y_fill, 'lightblue', label = 'Search space')

ax.plot(x, y1, label = r'$x_1 + x_2 \leq 4$')
ax.axhline(3, color = 'orange', linestyle = '--', label = r"$x_2 \leq 3$")
ax.axvline(2, color = 'green', linestyle = '--', label = r"$x_1 \leq 2$")

if model.success:
  opt_x1, opt_x2 = model.x
  ax.plot(opt_x1, opt_x2, 'ro', label = 'Optimal solution')
  ax.text(opt_x1 + 0.2, opt_x2 + 0.2, f'Opt ({opt_x1:.2f}, {opt_x2:.2f})\nMax z = {-model.fun:.2f}')

  ax.set_xlim(0, 5)
  ax.set_ylim(0, 5)
  ax.set_xlabel(r'$x_1$')
  ax.set_ylabel(r'$x_2$')
  ax.set_title('Simplex algorithm: search space and the optimal solution')
  ax.grid(True)
  ax.legend()

  fig, model.x, -model.fun
```

# Task 1:

*   Maximize the given function with the Simplex algorithm
    *   $ z = 4 * x_1 + 3 * x_2 $
*   Subject to:
    *   $2*x_1 + x_2 \leq 8$
    *   $x_1 + x_2 \leq 6$
    *   $x_1, x_2 \ge 0$
* Visualize the feasible region of the solutions and mark the optimal solution in it





```python
C = [-4, -3]
A = [[2, 1], [1, 1]]
B = [8, 6]

bounds = [(0, None), (0, None)]

model = linprog(C, A_ub = A, b_ub = B, bounds = bounds, method = 'simplex')

fig, ax = plt.subplots(figsize = (6, 6))

x = np.linspace(0, 10, 400)
y1 = (8 - 2*x)
y2 = (6 - x)

y1 = np.clip(y1, 0, 10)
y2 = np.clip(y2, 0, 10)

ax.plot(x, y1, label = r'$2x_1 + x_2 \leq 8$')
ax.plot(x, y2, label = r'$x_1 + x_2 \leq 6$')

x_fill = [0, 0, 2, 4, 0]
y_fill = [0, 6, 4, 0, 0]

ax.fill(x_fill, y_fill, 'lightblue', alpha = 0.5, label = 'feasible region')

if model.success:
  x1, x2 = model.x
  ax.plot(x1, x2, 'ro', label = 'Optimal solution')
  ax.text(x1 + 0.2, x2 + 0.2, f'({x1:.2f}, {x2:.2f})\n Maximum z = {-model.fun:.2f})')
  ax.set_xlabel(r'$x_1$')
  ax.set_ylabel(r'$x_2$')
  ax.set_title('Task 1')
  ax.grid(True)
  ax.legend()
  plt.show()
```

# Task 2

*   Maximize:
    * $z = 5 * x_1 + 4 * x_2$  
*   Subject to:
    *   $6*x_1 + 4*x_2 \leq 24$
    *   $x_1 + 2*x_2 \leq 6$
    *   $-x1 + x2 \leq 1$
    *   $x_1, x_2 \ge 0$
*   Visualize the feasible region of the solutions and mark the optimal solution in it

```python
from sympy import symbols, Eq, solve
c = [-5, -4]

A = [[6, 4], [1, 2], [-1, 1]]
b = [24, 6, 1]

bounds = [(0, None), (0, None)]

model = linprog(c, A_ub = A, b_ub = b, bounds = bounds, method = 'simplex')

x = np.linspace(0, 10, 400)

y1 = (24 - 6 * x) / 4
y2 = (6 - x) / 2
y3 = x + 1

x1, x2 = symbols('x1 x2')

sol1 = solve([Eq(6*x1 + 4*x2, 24), Eq(x1 + 2*x2, 6)], (x1, x2))
sol2 = solve([Eq(x1 + 2*x2, 6), Eq(-x1 + x2, 1)], (x1, x2))
sol3 = solve([Eq(6*x1+4*x2, 24), Eq(-x1 + x2, 1)], (x1, x2))

vertices = np.array([
    [sol1[x1], sol1[x2]],
    [sol2[x1], sol2[x2]],
    [sol3[x1], sol3[x2]]]
)

fig, ax = plt.subplots(figsize = (6, 6))

ax.plot(x, y1, 'b', label = r'$6x_1+4x_2 \leq 24$')
ax.plot(x, y2, 'orange', label = r'$x_1 +2x_2 \leq 6$')
ax.plot(x, y3, 'g', label = r'$-x_1+x_2 \leq 1$')

x_vals, y_vals = np.meshgrid(np.linspace(0, 10, 400), np.linspace(0, 10, 400))

mask = ((6*x_vals + 4*y_vals <= 24) &
        (x_vals + 2*y_vals <= 6) &
        (-x_vals + y_vals <= 1)).astype(int)

ax.contourf(x_vals, y_vals, mask, levels = [0.5, 1.5], color = 'cyan', alpha = 0.3)
ax.plot(vertices[:, 0], vertices[:, 1], 'ko', label = 'Vertices')

if model.success:
    x1_opt, x2_opt = model.x
    # print(x1_opt, x2_opt)
    ax.plot(x1_opt, x2_opt, 'ro', label = 'Optimal point')
    ax.text(x1_opt + 0.2, x2_opt + 0.2, f'O:({x1_opt:.2f}, {x2_opt:.2f})\n Maximum z = {-model.fun:.2f}')

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_xlabel(r'x_1')
ax.set_ylabel(r'x_2')
ax.set_title('Task 2')
ax.grid(True)
ax.legend()
plt.show()
```
