* create a function that accepts *argv. The first two arguments should be returned as one array, and the second two as another array.

```python
def first_task(*argv):
  return [argv[0:2], argv[2:4]]
```

# Task 2:
* write a function that calculates n squares of numbers that takes only one argument
* e.g. second_task(4) returns \[0, 1, 4, 9]

```python
def squares(limit):
  r = 0
  ret = []

  while r < limit:
    ret.append(r**2)
    r += 1

  return ret
```

```python
squares(4)
```

```python
for i in squares(4):
  print(i)
```
